```python
figsize(20,10)
for i, phi in enumerate([0.5]):
    figure(i)
    for eta in [0.1]:
        for rule in ["weighted", "unweighted"]:
            try:
                ts = loadtxt('data/ts_phi{0}_eta{1}_{2}.dat'.format(phi, eta, rule))
            except:
                continue
            #ts = ts * (1.0 - phi)
            runs = len(ts)
            ts.sort()

            c = 1.0/runs
            ecdf = []
            for t in ts:
                ecdf.append(c)
                c += 1.0/runs
            
            semilogx(ts,ecdf,label='$\phi={0},\eta={1}$'.format(phi,eta)) 
    
        axis('tight')
        axis([1, 1000, None, None])
        legend(loc='lower right')
        xlabel('t')
        ylabel('ECDF')
        savefig('fig/ecdf_phi{0}.pdf'.format(phi))
```
